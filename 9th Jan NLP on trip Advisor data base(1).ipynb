```python
def text_process(mess):            ### creating a function
    """                                                        ## a docstring
    1. remove the punctuation
    2. remove the stopwords
    3. return the list of clean textwords
    
    """
    nopunc = [char for char in mess if char  not in string.punctuation]  #focus is on the 'not'
    nopunc = "".join(nopunc)
    
    return [ word for word in nopunc.split() if word not in stopwords.words("english")]  #focus is on 'not'
```

```python
example  = ("this is just and example ! to show ^ to remove @ the punctions and specail character")
```

```python
text_process(example)
```

```python
trip.Review= trip.Review.str.lower()
trip.Review
```

```python
trip['Review'].apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer = CountVectorizer(analyzer= text_process).fit(trip['Review'])
```

```python
bow_transformer.vocabulary_
```

```python
trip_bow = bow_transformer.transform(trip['Review'])
```

```python
from sklearn.model_selection import train_test_split
df_x_train , df_x_test , df_y_train , df_y_test = train_test_split(trip_bow ,trip.Rating , test_size= .2 )
```

```python
print(df_x_train.shape)
print(df_y_train.shape)
print("---------------")
print(df_x_test.shape)
print(df_y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(df_x_train , df_y_train)
```

```python
pred = nb.predict(df_x_test)
```

```python
from sklearn.metrics import confusion_matrix
tab = confusion_matrix(df_y_test , pred)
tab
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(df_y_test ,pred )
```
